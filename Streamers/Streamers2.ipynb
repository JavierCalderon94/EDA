# OBTENCIÓN DE DATAFRAME DE STREAMS DEL ÚLTIMO MES DE CADA STREAMER

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
import time
import random

datos_streamers = pd.read_csv("1000_Streamers_2.csv")
nombre = list(datos_streamers.iloc[:,0].values)

dict={"streamer":[],
      "fecha":[],
     "duracion":[],
     "media viewers":[],
     "max viewers":[],
     "followers":[],
     "contenido 1":[],
     "contenido 2":[],
     "contenido 3":[],
     "contenido 4":[]}

lista_fallos=[]

for num, streamer in enumerate(nombre):
    if num==20:
        break
    try:
        service = Service(executable_path='./chromedriver.exe')

        chrome_options = Options()
        chrome_options.add_argument("--headless")
        driver = webdriver.Chrome(service=service, options=chrome_options)

        url = "https://twitchtracker.com"+datos_streamers.iloc[num,1]+"/streams"

        print("Obteniendo datos de:",streamer)
        driver.get(url)

        personaldata = driver.find_element(By.XPATH, '/html/body/div[4]/div[2]/div[1]/div[3]/div[2]/button[1]/p')
        personaldata.click()

        soup = BeautifulSoup(driver.page_source, "html.parser")

        pares = []
        impares = []
        contador = 0

        for i in soup.find_all("tr", class_="even"):
            for h in i.find_all("span"):
                if h.get_text()[0]=="+":
                    continue
                else:
                    pares.append(h)

        for i in soup.find_all("tr", class_="odd"):
            for h in i.find_all("span"):
                if h.get_text()[0]=="+":
                    continue
                else:
                    impares.append(h)

        for h,j in enumerate(impares):
            
            contador+=1
            match contador:
                case 1:
                    dict["fecha"].append(j.get_text())
                    dict["fecha"].append(pares[h].get_text())
                case 2:
                    dict["duracion"].append(j.get_text())
                    dict["duracion"].append(pares[h].get_text())
                case 3:
                    dict["media viewers"].append(j.get_text())
                    dict["media viewers"].append(pares[h].get_text())
                case 4:
                    dict["max viewers"].append(j.get_text())
                    dict["max viewers"].append(pares[h].get_text())
                case 5:
                    dict["followers"].append(j.get_text())
                    dict["followers"].append(pares[h].get_text())
                case 6:
                    contador=0

        contador = 0
        for i in soup.find_all("td", class_="games hidden-xs hidden-sm"):

            match contador:
                case 1: 
                    dict["contenido 2"].append(pd.NA)
                    dict["contenido 3"].append(pd.NA)
                    dict["contenido 4"].append(pd.NA)
                case 2: 
                    dict["contenido 3"].append(pd.NA)
                    dict["contenido 4"].append(pd.NA)
                case 3: 
                    dict["contenido 4"].append(pd.NA) 

            contador = 0      

            for h, j in enumerate(i.find_all("img")):
                contador+=1
                
                match h:
                    case 0:
                        dict["contenido 1"].append(j.get("data-original-title"))
                    case 1:
                        dict["contenido 2"].append(j.get("data-original-title"))
                    case 2:
                        dict["contenido 3"].append(j.get("data-original-title"))
                    case 3:
                        dict["contenido 4"].append(j.get("data-original-title"))

        match contador:
            case 1: 
                dict["contenido 2"].append(pd.NA)
                dict["contenido 3"].append(pd.NA)
                dict["contenido 4"].append(pd.NA)
            case 2: 
                dict["contenido 3"].append(pd.NA)
                dict["contenido 4"].append(pd.NA)
            case 3: 
                dict["contenido 4"].append(pd.NA)

        #--------------------------------------------------------------------------------------------------------------------------------------------------
        if (len(pares)+len(impares))==120:
            while soup!=1:
                try:
                    pagina_2 = driver.find_element(By.XPATH,'/html/body/div[2]/div[4]/div[5]/div[2]/div/ul/li[2]/a')
                    pagina_2.click()
                    break
                except:
                    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                    continue

            soup = BeautifulSoup(driver.page_source, "html.parser")
            pares = []
            impares = []
            contador = 0

            for i in soup.find_all("tr", class_="even"):
                for h in i.find_all("span"):
                    if h.get_text()[0]=="+":
                        continue
                    else:
                        pares.append(h)

            for i in soup.find_all("tr", class_="odd"):
                for h in i.find_all("span"):
                    if h.get_text()[0]=="+":
                        continue
                    else:
                        impares.append(h)

            for h,j in enumerate(impares):
                contador+=1
                match contador:
                    case 1:
                        dict["fecha"].append(j.get_text())
                        dict["fecha"].append(pares[h].get_text())
                    case 2:
                        dict["duracion"].append(j.get_text())
                        dict["duracion"].append(pares[h].get_text())
                    case 3:
                        dict["media viewers"].append(j.get_text())
                        dict["media viewers"].append(pares[h].get_text())
                    case 4:
                        dict["max viewers"].append(j.get_text())
                        dict["max viewers"].append(pares[h].get_text())
                    case 5:
                        dict["followers"].append(j.get_text())
                        dict["followers"].append(pares[h].get_text())
                    case 6:
                        contador=0

            contador = 0
            for i in soup.find_all("td", class_="games hidden-xs hidden-sm"):

                match contador:
                    case 1: 
                        dict["contenido 2"].append(pd.NA)
                        dict["contenido 3"].append(pd.NA)
                        dict["contenido 4"].append(pd.NA)
                    case 2: 
                        dict["contenido 3"].append(pd.NA)
                        dict["contenido 4"].append(pd.NA)
                    case 3: 
                        dict["contenido 4"].append(pd.NA) 

                contador = 0      

                for h, j in enumerate(i.find_all("img")):
                    contador+=1
                    
                    match h:
                        case 0:
                            dict["contenido 1"].append(j.get("data-original-title"))
                        case 1:
                            dict["contenido 2"].append(j.get("data-original-title"))
                        case 2:
                            dict["contenido 3"].append(j.get("data-original-title"))
                        case 3:
                            dict["contenido 4"].append(j.get("data-original-title"))

            match contador:
                case 1: 
                    dict["contenido 2"].append(pd.NA)
                    dict["contenido 3"].append(pd.NA)
                    dict["contenido 4"].append(pd.NA)
                case 2: 
                    dict["contenido 3"].append(pd.NA)
                    dict["contenido 4"].append(pd.NA)
                case 3: 
                    dict["contenido 4"].append(pd.NA)

            driver.close()

        resta = len(dict["fecha"]) - len(dict["streamer"])
        for i in list(range(resta)):
            dict["streamer"].append(streamer)

        time.sleep(random.randint(2,6))

    except:
        print("fallo")
        lista_fallos.append("streamer")
```
