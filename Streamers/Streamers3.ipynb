# PARTE 2, OBTENER CONTENIDO/JUEGO MÁS RELEVANTE DEL CANAL

```python
from fake_useragent import UserAgent
import requests
import pandas as pd
from datetime import datetime
from bs4 import BeautifulSoup as bs
import time
import random

ua = UserAgent()
headers = {'User-Agent': ua.random}

```

```python
datos_streamers = pd.read_csv("1000_Streamers.csv")
```

```python
datos_streamers.iloc[0,1]
```

```python
nombre = list(datos_streamers.iloc[:,0].values)
```

```python
for i, j in enumerate(nombre):

    try:
        url = "https://twitchtracker.com"+datos_streamers.iloc[i,1]+"/games"
        response = requests.get(url, headers=headers)
        soup = bs(response.content, 'html.parser')
        datos_streamers.loc[datos_streamers["streamer"]==j,"contenido principal"] = soup.find_all("td")[2].get_text()
        print("Scrapeo n",i,"Obteniendo datos de ",j)
        time.sleep(random.randint(3,6))
    except:
        break
```

```python
datos_streamers.to_csv("1000_Streamers_2.csv", index=False)
```
