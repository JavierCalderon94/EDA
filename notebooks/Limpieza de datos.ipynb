```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Obtenidos 3 csv, el primero con los datos de 1000 streamer, el segundo y el tercero con los datos de sus últimos 40 streams(en caso de haberlos hecho). Estos dos últimos están separados ya que el scrapping dió fallos y tuve que hacer un segundo intento únicamente con los que habían dado fallo, vamos a juntarlos.

```python
datos_streams = pd.read_csv("DATOS_STREAMS_PPAL.csv", index_col=0)
datos_streams
```

```python
datos_streams2 = pd.read_csv("DATOS_STREAMS.csv", index_col=0)
datos_streams2
```

```python
streams= pd.concat([datos_streams,datos_streams2])
streams
```

Ahora al estar unidos no están en orden, vamos a obtener los índices del dataframe principal y a ordenarlo por este valor y despues por los indices que están en esta tabla

```python
top_1000_streamers = pd.read_csv("data/1000_streamers_3.csv")
top_1000_streamers.reset_index(inplace=True)
top_1000_streamers
```

```python
streamsf = pd.merge(streams, top_1000_streamers[['streamer', 'index']], on='streamer', how='left')
streamsf
```

```python
streamsf.rename(columns={"index":"Borrar1"},inplace=True)
streamsf.reset_index(inplace=True)
```

```python
streamsf.sort_values(by=["Borrar1","index"], inplace=True)
```

```python
streamsf.drop(["Borrar1","index"],axis=1,inplace=True)
streamsf
```

Ahora aprovechamos para hacer limpieza de datos antes de volver a guardarlo en csv

```python
streamsf.rename(columns={"duracion":"duracion(horas)"}, inplace=True)
```

```python
streamsf["duracion(horas)"] = streamsf["duracion(horas)"].replace({' hrs':''}, regex=True)
```

```python
streamsf[["media viewers", "max viewers", "followers"]] = streamsf[["media viewers", "max viewers", "followers"]].replace({",":''}, regex=True)
```

```python
streamsf["fecha"] = streamsf["fecha"].replace({"Jan":"01","Feb":"02","Mar":"03","Apr":"04","May":"05","Jun":"06","Jul":"07","Aug":"08","Sep":"09","Oct":"10","Nov":"11","Dec":"12"}, regex=True)
```

```python
streamsf = pd.merge(streamsf, top_1000_streamers[['streamer', 'contenido principal']], on='streamer', how='left')
```

```python
streamsf["resumen"]=0
```

```python
streamsf.loc[(streamsf["contenido 1"]==streamsf["contenido principal"])&(streamsf["contenido 2"].isnull()),"resumen"]="Propio"
```

```python
streamsf.loc[(streamsf["contenido 1"]!=streamsf["contenido principal"])&(streamsf["contenido 2"].isnull()),"resumen"]="Otro contenido"
```

```python
streamsf.loc[((streamsf["contenido 1"]==streamsf["contenido principal"])&(~streamsf["contenido 2"].isnull()))|(streamsf["contenido 2"]==streamsf["contenido principal"])|(streamsf["contenido 3"]==streamsf["contenido principal"])|(streamsf["contenido 4"]==streamsf["contenido principal"]),"resumen"]="Propio + otros"
```

```python
streamsf.loc[(streamsf["contenido 1"]!=streamsf["contenido principal"])&(streamsf["contenido 2"]!=streamsf["contenido principal"])&(streamsf["contenido 3"]!=streamsf["contenido principal"])&(streamsf["contenido 4"]!=streamsf["contenido principal"])&(~streamsf["contenido 2"].isnull()),"resumen"]="Otro contenido o más"
```

```python
streamsf["fecha"] = streams["fecha"].replace({"Jan":"01"}, regex=True)
streamsf["fecha"] = streams["fecha"].replace({"Feb":"02"}, regex=True)
streamsf["fecha"] = streams["fecha"].replace({"Mar":"03"}, regex=True)
streamsf["fecha"] = streams["fecha"].replace({"Apr":"04"}, regex=True)
streamsf["fecha"] = streams["fecha"].replace({"May":"05"}, regex=True)
streamsf["fecha"] = streams["fecha"].replace({"Jun":"06"}, regex=True)
streamsf["fecha"] = streams["fecha"].replace({"Jul":"07"}, regex=True)
streamsf["fecha"] = streams["fecha"].replace({"Aug":"08"}, regex=True)
streamsf["fecha"] = streams["fecha"].replace({"Sep":"09"}, regex=True)
streamsf["fecha"] = streams["fecha"].replace({"Oct":"10"}, regex=True)
streamsf["fecha"] = streams["fecha"].replace({"Nov":"11"}, regex=True)
streamsf["fecha"] = streams["fecha"].replace({"Dec":"12"}, regex=True)
```
